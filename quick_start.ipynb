# Fast Photo Mosaic

## Load Image Directories

```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
from photo_mosaic import *
%matplotlib inline
types = ["manmade","natural"]
```

## Process Source Images

```python
for type in types:
    data_augmentation(type)
with open(DATA_pth+types[0]+".txt","r") as f_txt:
    data_list = f_txt.read().splitlines()
```

## Choose a Target Image

```python
img_target = cv2.imread(data_list[199])
plt.imshow(img_target[:,:,::-1])
plt.show()
```

## Create Composite Image
Note: the pool need only to be built once before restarting kernel

```python
tiles = (32,32)
img_composite = get_composite(img_target,data_list,tiles,type=types[0])
```

### Raw Composition

```python
# raw composite
plt.figure(figsize=(15,15))
plt.subplot(1,2,1)
plt.imshow(img_composite[:,:,::-1])
plt.subplot(1,2,2)
plt.imshow(img_target[:,:,::-1])
plt.show()
```
